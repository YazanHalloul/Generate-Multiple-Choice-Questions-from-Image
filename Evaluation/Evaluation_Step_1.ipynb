# Declare Final prompt that get best response

```python
text ="""
Describe the image with specific details, focusing on identifiable objects, their positions, colors, and actions. Avoid using subjective language or broad descriptions. Include enough detail to allow someone to form precise questions about the image.    """

prompt = """
Describe the image with specific details, focusing on the following key elements:
  1. Identifiable Objects:
     - List and describe all objects in the image, including their attributes such as size, shape, and color.
  2. Positions and Relationships:
     - Specify the positions of objects relative to each other and describe spatial relationships and interactions.
  3. Colors and Textures:
     - Note the colors of objects and any notable textures or patterns.
  4. Actions and Movements:
     - Describe any actions being performed and details about any motion.
  5. Expressions and Emotions:
     - For people or animals, describe facial expressions or emotional states.
  6. Context and Setting:
     - Provide details about the setting or environment and the overall scene context.
  7. Clothing and Accessories:
     - Describe what the people in the image are wearing, including specific items and colors.
  8. Lighting and Atmosphere:
     - Note the lighting conditions, their effect on the scene, and describe the general atmosphere or mood.

- Avoid using subjective language or broad descriptions. Include enough detail to allow someone to form precise questions about the image."""
```

# Load VQA model

```python
import torch
from PIL import Image
from transformers import AutoModel, AutoTokenizer

model_MiniCPM = AutoModel.from_pretrained('openbmb/MiniCPM-Llama3-V-2_5-int4', trust_remote_code=True)
tokenizer = AutoTokenizer.from_pretrained('openbmb/MiniCPM-Llama3-V-2_5-int4', trust_remote_code=True)
```

# Insert Generated Description to CSV file

```python
import pandas as pd
import requests
from PIL import Image
from io import BytesIO

msgs = [{'role': 'user', 'content': text}]

# Load the existing CSV file into a DataFrame
df = pd.read_csv('stanford_df_rectified.csv')

# Add headers for the new columns
if 'Generated_description_1' not in df.columns:
    df['Generated_description_1'] = ''
if 'Generated_description_2' not in df.columns:
    df['Generated_description_2'] = ''

for index, row in df.iterrows():
    if index >= 950:  # Stop after processing the 950 row (0-based index)
        break

    image_url = row['url']
    description = row['Paragraph']

    # Fetch and open the image
    response = requests.get(image_url)
    img = Image.open(BytesIO(response.content)).convert('RGB')

    # Generate descriptions using MiniCPM-Llama3-V-2_5 model
    res1 = model_MiniCPM.chat(
        image=img,
        msgs=msgs,
        tokenizer=tokenizer,
        sampling=True,
        temperature=0.5,
    )

    res2 = model_MiniCPM.chat(
        image=img,
        msgs=msgs,
        tokenizer=tokenizer,
        sampling=True,
        temperature=0.4,
    )

    # Save the generated descriptions back to the DataFrame
    df.at[index, 'Generated_description_1'] = res1
    df.at[index, 'Generated_description_2'] = res2
    print(f"Processed row {index + 1}")

# Save the DataFrame back to a CSV file with the changes
df.to_csv('stanford_df_rectified.csv', index=False)
```
